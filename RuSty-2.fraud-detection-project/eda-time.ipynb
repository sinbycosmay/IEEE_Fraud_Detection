```python
v = ['V' + str(i) for i in range(1, 340)]
v
```

```python
df[v]
```

```python
dfv = df[v]
dfv
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

pca_unscaled = pca.fit_transform(dfv.fillna(0))
```

```python
pca_unscaled = pd.DataFrame( pca_unscaled , columns=['lol1', 'lol2'])
```

```python
sns.relplot( x = 'lol1', y = 'lol2', data=pca_unscaled )
```

```python
pca2 = PCA(n_components = 2)
pca_scaled = pca2.fit_transform(dfv_scaled)
```

```python
m = ['M' + str(i) for i in range(1, 10)]
m
```

```python
df[m]
```

```python
m_dict = {
    'T' : 1, 'F' : 0, 'M0' : 0, 'M1' : 1, 'M2' : 2, np.nan : 0
}
```

```python
dfm = df[m].replace(m_dict)
dfm
```

```python
ss = StandardScaler()
dfm = ss.fit_transform(dfm)
```

```python
pca3 = PCA(n_components=2)
dfm = pca3.fit_transform(dfm)
```

```python
dfm = pd.DataFrame(dfm, columns=['lol1', 'lol2'])
dfm['isFraud'] = df.isFraud
```

```python
sns.relplot( x = 'lol1', y = 'lol2', hue='isFraud', data=dfm )
```
